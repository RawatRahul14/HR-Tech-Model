```python
import pandas as pd
from io import StringIO
```

```python
df = pd.read_csv("Data/cleaned_df.csv")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, Input, Dense
from tensorflow.keras.models import Model
import numpy as np
import pandas as pd
```

```python
vectorizer = tf.keras.layers.TextVectorization()
vectorizer.adapt(df['skills'])
```

```python
sequences = vectorizer(df['skills']).numpy()
```

```python
vocab_size = len(vectorizer.get_vocabulary())
embedding_dim = 50  # Number of dimensions for embedding vectors
```

```python
from tensorflow.keras.layers import Embedding, Input, Dense, GlobalAveragePooling1D, Dropout
```

```python
inputs = Input(shape=(None,), dtype=tf.int32)
x = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequences.shape[1])(inputs)
x = GlobalAveragePooling1D()(x)

# Add hidden layers
x = Dense(64, activation='relu')(x)  # First hidden layer
x = Dropout(0.5)(x)  # Dropout layer for regularization
x = Dense(32, activation='relu')(x)  # Second hidden layer

# Output layer
outputs = Dense(1, activation='sigmoid')(x)  # Dummy output layer for example
```

```python
model = Model(inputs, outputs)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
targets = np.random.randint(2, size=(sequences.shape[0], 1))
```

```python
model.fit(sequences, targets, epochs=5, batch_size=128)
```

```python
model.summary()
```

```python
embedding_layer = model.get_layer(index=1)  # Embedding layer is the first layer
embedding_weights = embedding_layer.get_weights()[0]
```

```python
vocab = vectorizer.get_vocabulary()
for index, token in enumerate(vocab):
    if index < len(embedding_weights):  # Ensure index is within bounds
        print(f"Token: {token}, Embedding: {embedding_weights[index]}")
```
