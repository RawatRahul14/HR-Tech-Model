```python
df = df.drop('Contact Person', axis=1)
df.drop('Contact', axis=1, inplace=True)
df.drop('Benefits', axis=1,inplace=True)
df.head()
```

**creating a ew column called inumum experience required**

```python
import pandas as pd

def split_experience(experience):
    min_exp, max_exp = experience.split(' to ')
    min_exp = int(min_exp)
    max_exp = int(max_exp.split(' ')[0])  # Split by space to remove 'Years' and convert to int
    return pd.Series([min_exp, max_exp])


df[['Minimum Experience', 'Maximum Experience']] = df['Experience'].apply(split_experience)
```

```python
df['Average Experience'] = (df['Minimum Experience'] + df['Maximum Experience']) / 2
```

**Calculating minimum salary, maximum salary and average salary**

```python
def split_salary(salary_range):
  min_salary, max_salary = salary_range.split('-')
  min_salary = (min_salary)
  max_salary = (max_salary)
  return pd.Series([min_salary, max_salary])

df[['Minimum Salary', 'Maximum Salary']] = df['Salary Range'].apply(split_salary)
```

```python
df['Minimum Salary'] = df['Minimum Salary'].replace({'\$': '', 'K': '', ',': ''}, regex=True).astype(int)
df['Maximum Salary'] = df['Maximum Salary'].replace({'\$': '', 'K': '', ',': ''}, regex=True).astype(int)
```

```python
df['Average Salary'] = (df['Minimum Salary'] + df['Maximum Salary']) / 2
```

**#extracting year from job posting date**

```python
# Convert 'Job Posting Date' to datetime format
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
df['JobPostingYear'] = df['Job Posting Date'].dt.year
```

**categories for company size**

```python
bins = [0, 10000, 50000, 100000, 10000000]
labels = ['small', 'medium', 'large', 'enterprise']

df['Company_size'] = pd.cut(df['Company Size'], bins=bins, labels=labels, right=True, include_lowest=True)
```

**one hot encoding on preference column**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
encoded_df = ohe.fit_transform(df[['Preference']])
#converting it to a pandas dataframe because can oly concatinate dataframe and series object but not rray
encoded_df = pd.DataFrame(encoded_df, columns=ohe.get_feature_names_out(['Preference']))
encoded_df.index = df.index
df= pd.concat([df, encoded_df], axis=1)
df.drop('Preference', axis=1, inplace=True)
df.drop('Preference_Male', axis = 1, inplace=True) #to avoid multicolinearity
df.head(2)
```
